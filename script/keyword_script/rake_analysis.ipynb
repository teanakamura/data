```python
from os.path import expanduser
home = expanduser('~')
import numpy as np

source1 = home + '/Data/cnndm-pj/rake/doc/test.txt'
source2 = home + '/Data/cnndm-pj/cnndm/test.sum'
```

```python
R = 100
def count(rake, summ):
    rake_list = [ph.split() for ph in rake.split('<ph>')]
    summ_set = set(summ.split())
    for i, rake_elem in enumerate(rake_list):
        if i >= R: break
        count.length[i] += len(rake_elem)
        for w in rake_elem:
            if (w in summ_set):
                count.count[i] += 1
```

```python
N = 10000
ph_len = 0
count.count = [0] * R
count.length = [0] * R
with open(source1) as rake_f, open(source2) as summ_f:
    for i, (rake, summ) in enumerate(zip(rake_f, summ_f)):
        ph_len += len(rake.split('<ph>'))
        count(rake, summ)
        if i > N: break

count.count = np.array(count.count) / N
count.length = np.array(count.length) / N
print(count.count)
print(count.length)
print(ph_len / N)
```

```python
res = [r/l for r, l in zip(count.count, count.length)]
res
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(res)
plt.ylim(0, 0.25)
plt.show()

plt.plot(count.count)
plt.ylim(0, 1.2)
plt.show()

plt.plot(count.length)
plt.ylim(0, 6)
plt.show()
```
